#Task 1
Take a string of integers seperated by spaces. Then create a dictionary with the first two digits as unique keys and the rest of the number as the value of the key.

**Sample Input:**

17101454  18303035  21210031  11505060  16162050  19101154  17126524

**Output:**

{17: [101454,126524], 18: [303035], 21: [210031], 16: [162050], 19: [101154] }

```python
inp= input().split(" ")
dict_1={}

for i in inp:
  dict_1[int(i[0:2])] = [] 

for i in inp:
  if (int(i[0:2]) in dict_1.keys()):
    dict_1[int(i[0:2])].append(int(i[2:]))

print(dict_1)
```

#Task 2

Take a string containing cgpa of a student and the course fees. If the cgpa is equal or grater than 3.95, 75% scholarship granted on the course fees. If cgpa grater than 3.85, 50% scholarship is granted. Otherwise no scholarship is granted. Print the new amounts of the course fees and total waived amount.

**Sample Input #1:**

3.92,16000,12000,16000,8000

**Output:**

[8000,6000,8000,4000]

Total Waived: 26000


**Sample Input #2:**

3.98,16000,12000,16000,8000

**Output:**

[4000,3000,4000,2000]

Total Waived: 39000

```python
info= input().split(",")
course_fee=[]

for i in range (1,len(info)):
  course_fee.append(int(info[i]))

waived_fee=[]
for i in course_fee:
  if (float(info[0])>= 3.95):
    waived_fee.append(i*0.25)
  elif (float(info[0])> 3.85):
    waived_fee.append(i*0.5)
  else:
    waived_fee.append(i)

total_waived=0
for i in range (0, len(course_fee)):
  total_waived+= course_fee[i]- waived_fee[i]


print(waived_fee)
print(total_waived)
```

#Task 4
Take a dictionary, find the total ASCII values of the keys and reorder the dictionary in ascending order depending on the total ASCII values.


Sample Dictionary:

{"Morpheous": [77], "Neo":[1], "Trinity": [10,25], "Niobe": [41,54]}

Output:

{"Neo":[1], "Niobe": [41,54], "Trinity": [10,25], "Morpheous": [77]}

```python
#Shabbash Muntasir
my_dict={"Morpheous":[77], "Neo":[1], "Trinity":[10,25], "Niobe":[41,54]}
key_list=[]
for i in my_dict.keys():
    key_list.append(i)
asc_li=[]
for i in key_list:
    sum_asc=0
    for j in i:
        sum_asc+=ord(j)
    asc_li.append(sum_asc)
for i in range(0,len(asc_li)-1):
    for j in range(0,len(asc_li)-1):
        if asc_li[j]>asc_li[j+1]:
            temp_num=asc_li[j]
            temp_key=key_list[j]
            asc_li[j]=asc_li[j+1]
            key_list[j]=key_list[j+1]
            asc_li[j+1]=temp_num
            key_list[j+1]=temp_key
new_dict={}
for i in key_list:
    new_dict[i]= my_dict[i]
print(new_dict)
```

#Task 5
Consider the following dictionary containing total marks of a student:

**Alex= {"Final" : 33, "Mid" 18, "Lab": 16, "Assignments":[10,8,8.5,4,9]}**

Calculate the total marks considering the average of the best n-2 Assignments if number of Assignments is grater than 4, otherwise consider best n-1.

Output: 

Total Marks: 76

```python
Alex= {"Final" : 33, "Mid" :18, "Lab": 16, "Assignments":[10,8,8.5,4,9]}

total= Alex["Final"]+Alex["Mid"]+Alex["Lab"]

Alex["Assignments"].sort()
Alex["Assignments"].reverse()

print(Alex["Assignments"])

total_a=0
if (len(Alex["Assignments"]))>4:
    for i in range(0, len(Alex["Assignments"])-2):
      total_a+= Alex["Assignments"][i]
    total+= total_a/(len(Alex["Assignments"])-2)
else:
    for i in range(0, len(Alex["Assignments"])-1):
      total_a+= Alex["Assignments"][i]
    total+= total_a/(len(Alex["Assignments"])-1)

print(total)
```

#Task 7

Imagine you are tasked with creating g-suite email for the students. Now take the first, last name and student id in the following way to create two g-suite emails.

student= ["Sifat",2,"Tanvir",3,17101454,4]

fat.vir.1454@g.bracu.ac.bd
