```python
#1
class Student():
    ID=0
    def __init__(self,name,dept,age,cgpa):
        Student.ID+=1
        self.name=name
        self.department=dept
        self.age=age
        self.cgpa=cgpa
    
    @classmethod
    def from_String(cls,info):
        name,dept,age,cgpa=info.split("-")
        
        return cls(name,dept,age,cgpa)

    def get_details(self):
        print('ID:',Student.ID)
        print('Name:',self.name)
        print('Department:',self.department)
        print('Age:',self.age)
        print('CGPA:',self.cgpa)
s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
print("Instance variable assigns values inside the class methods and for each objects instance variables are "
      "different.\nOn the other hand, class variable assigns a value in the class decoration and these variables are" 
      "shared amongst objects of the class.")
print("Instance methods accesses the attributes through the self parameter.\nOn the other hand, class methods accesses "
      "the class through cls parameter.")
```

```python
#2
class Assassin():
    count=0
    def __init__(self,name,rate):
        self.name=name
        self.successRate=rate
        Assassin.count+=1

    @classmethod
    def failureRate(cls,name,rate):
        rate=100-rate
        return cls(name,rate)

    @classmethod
    def failurePercentage(cls,name,rate):
        x=''
        for i in range(len(rate)):
            if rate[i]=='%':
                break
            else:
                x=x+rate[i]
        rate=100-int(x)
        return cls(name,rate)
    def printDetails(self):
        print(f'Name: {self.name}')
        print(f'Success rate: {self.successRate}%')
        print(f'Total number of Assassin: {Assassin.count}')
        
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
#5
class Employee():
    def __init__(self,name,period):
        self.name=name
        self.workingPeriod=period
    @classmethod
    def employeeByJoiningYear(cls,name,year):
        period=2021-year
        return cls(name,period)
    @classmethod
    def experienceCheck(cls,period,gender):
        if period<3 and gender=='male':
            return('He is not experienced')
        elif period<3 and gender=='female':
            return('She is not experienced')
        elif period>=3 and gender=='male':
            return('He is experienced')
        elif period>=3 and gender=='female':
            return('She is experienced')



employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#7
class Cat:
    Number_of_cats=0
    def __init__(self,color,action):
        self.color=color
        self.action=action
        Cat.Number_of_cats+=1
    @classmethod
    def no_parameter(cls):
        color,action='White','sitting'
        return cls(color,action)
    @classmethod
    def first_parameter(cls,color):
        color,action=color,'sitting'
        return cls(color,action)
    @classmethod
    def second_parameter(cls,action):
        color,action='White',action
        return cls(color,action)
    def changeColor(self,color):
        self.color=color
    def printCat(self):
        print(self.color,'cat is',self.action)


print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#8
class Cylinder:
    radius=5
    height=18
    def __init__(self,val1,val2):
        self.value1=val1
        self.value2=val2
        print('Default radius=',Cylinder.radius,'and height=',Cylinder.height)
        Cylinder.radius=self.value1
        Cylinder.height=self.value2
        print('Updated radius=',Cylinder.radius,'and height=',Cylinder.height)
    @staticmethod
    def area(num1,num2):
        import math
        area=(2*math.pi*float(Cylinder.radius)*float((Cylinder.radius)+float(Cylinder.height)))
        print('Area:',area)
    @staticmethod
    def volume(num1,num2):
        import math
        print("Volume:",(math.pi*(float(Cylinder.radius)**2)*float(Cylinder.height)))
    @classmethod
    def swap(cls,temp,new):
        return cls(new,temp)
    @classmethod
    def changeFormat(cls,info):
        radius,height=info.split("-")
        return cls(float(radius),float(height))
        


c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#10
class SultansDine:
    num_of_branch=0
    total_sell=0
    detail=[]
    def __init__(self,place):
        self.place=place
        self.quantity=None
        SultansDine.num_of_branch+=1
    
    def sellQuantity(self,quantity):
        if quantity < 10:
            branch_sell = quantity * 300
        elif quantity < 20:
            branch_sell =quantity * 350
        else:
            branch_sell = quantity * 400
        self.quantity=branch_sell
        SultansDine.total_sell+=branch_sell
        info=self.place+'*'+str(self.quantity)
        SultansDine.detail.append(info)

    def branchInformation(self):
        print(f'Branch Name: {self.place}')
        print(f'Branch Sell:{self.quantity}Taka')

    @classmethod
    def details(cl_name):
        print(f'Total Number of branch(s):{cl_name.num_of_branch}')
        print(f'Total Sell:{cl_name.total_sell}Taka')
        for i in cl_name.detail:
            info_li=i.split('*')
            print('Branch Name:',info_li[0],end=',')
            print("Branch Sell:",info_li[1],'Taka')
            br_sell=int(info_li[1])
            percent=br_sell/cl_name.total_sell*100
            round_percent=round(percent,2)
            print("Branch consists of total sell's:", round_percent,"%")
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()





```
